```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
student_data = pd.read_csv('StudentPerformanceFactors.csv')
student_data
```

```python
standard = StandardScaler()
impute_num = SimpleImputer(strategy='mean')
impute_cat = SimpleImputer(strategy='most_frequent')
label_encoder = LabelEncoder()
```

```python
student_data.describe(include='all')
```

```python
student_data.isnull().sum()
```

```python
y = student_data['Exam_Score']
X = student_data.drop('Exam_Score',axis=1)
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
num_col = X.select_dtypes(include='number').columns
cat_col = X.select_dtypes(include='object').columns
```

```python
for col in num_col:
    plt.figure()
    plt.title(f'{col} distribution')
    sns.boxplot(student_data[col])
    plt.show
```

```python
for col in cat_col :
  X_train[[col]] = impute_cat.fit_transform(X_train[[col]])  # Utiliser [[col]] pour maintenir le format DataFrame
  X_test[[col]] = impute_cat.transform(X_test[[col]])  # De mÃªme pour le test
```

```python
 for col in num_col:
    X_train[[col]] = standard.fit_transform(X_train[[col]])
    X_test[[col]] = standard.transform(X_test[[col]]) 
```

```python
for col in X_train.select_dtypes(include='number').columns:
    plt.figure()
    plt.title(f'{col} distribution')
    sns.boxplot(X_train[col])
    plt.show
```

```python
# Appliquer Ã  chaque colonne catÃ©gorielle
for col in cat_col:  # cat_col est une liste des colonnes catÃ©gorielles
    X_train[col] = label_encoder.fit_transform(X_train[col].astype(str))  # Transformer en string si nÃ©cessaire
    X_test[col] = label_encoder.transform(X_test[col].astype(str))  # Transformer en string si nÃ©cessaire
```

```python
cat_transformer = SimpleImputer(strategy='most_frequent')
```

```python
preprocessor = ColumnTransformer(
    transformers=
    [
        ('num', Pipeline(steps=[
            ('imputer',SimpleImputer(strategy='mean')),
            ('scaler',StandardScaler())
        ]), num_col),
        ('cat',cat_transformer,cat_col)
        
    ]
)
```

```python
my_pipeline = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',LinearRegression())
])
```

```python
my_pipeline.fit(X_train,y_train)
```

```python
my_pipeline.score(X_train, y_train)
```
