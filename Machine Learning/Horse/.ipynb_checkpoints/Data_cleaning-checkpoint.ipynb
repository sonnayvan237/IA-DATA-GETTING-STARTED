```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('horse.csv')
df
```

```python
df.describe()
```

```python
df['age'].isnull().sum()
```

```python
# compte le nombre de valeur manquante
for x in df.columns:
    print(f'{x}: {df[x].isnull().sum()}')
```

```python
# supprimer les colonnes avec plus de 100 valeurs manquantes
for y in df.columns :
    if df[y].isnull().sum() >= 100 :
        df.drop(y , axis = 1, inplace = True)
```

```python
for y in df.columns:
    # Vérifier si le type de la colonne est numérique (int64 ou float64)
    if df[y].dtypes == 'int64' or df[y].dtypes == 'float64':
        # Calculer la moyenne de la colonne et remplir les NaN
        df[y] = df[y].fillna(df[y].mean())
    else :
        # rempli les valeurs manquantes pour des types objets par le mode 
        df[y] = df[y].fillna(df[y].value_counts().idxmax())
```

```python
for x in df.columns:
    print(f'{x}: {df[x].isnull().sum()}')
```

```python
sns.boxplot(df['pulse'])
```

```python
# calcul de l'intervalle inter-quartile
q1 = df['pulse'].quantile(0.25)
print(q1)
q3 = df['pulse'].quantile(0.75)
print(q3)
IQR =q3-q1
print(IQR)
```

```python
# Affichage des outliers
bs = q3+(1.5*IQR)
bi = q1-(1.5*IQR)
outliers = [0]
for x in df['pulse']:
    if x < bi or x > bs :
        outliers.append(x)
outliers
```

```python
# standardisation avec le z_score
import numpy as np
mean = np.mean(df['pulse'])
std = np.std(df['pulse'])
outliers = [0]
for x in df['pulse']:
    z_score = (x - mean)/std
    outliers.append(z_score)
outliers  
sns.boxplot(outliers )
```
