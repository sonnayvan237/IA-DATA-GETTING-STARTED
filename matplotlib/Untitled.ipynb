```python
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt

data = {('Male' , 'Smoker') : 30,
       ('Male', 'No-smoker') : 20,
       ('Female', 'Smoker') : 15,
       ('Female', 'No-smoker') : 35}

mosaic(data, title='Gender vs Smoking Habits')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Générer des données aléatoires
np.random.seed(10)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# Créer le boxplot
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Exemple de Boxplot')
plt.xlabel('Groupe')
plt.ylabel('Valeurs')
plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic 
import matplotlib.pyplot as plt
plt.figure()
# Exemple de données
data = {('Male' , 'Smoker') : 30,
       ('Male', 'No-smoker') : 20,
       ('Female', 'Smoker') : 15,
       ('Female', 'No-smoker') : 35}

# Créer le mosaic plot
mosaic(data, title='Gender vs. Smoking Habits')
plt.show()
```
